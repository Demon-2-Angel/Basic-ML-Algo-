## K-Means Clustering Algorithm:

### Initialize:
1. Choose the number of clusters, `k`.
2. Randomly initialize `k` cluster centroids.

### Assign Data Points to Clusters:
1. For each data point, calculate the distance to each centroid.
2. Assign the data point to the cluster whose centroid is closest.

### Update Centroids:
1. Recalculate the centroid of each cluster as the mean of all data points assigned to that cluster.

### Repeat:
1. Repeat steps 2 and 3 until convergence:
    - No more changes in the assignment of data points to clusters.
    - Centroids do not change significantly.

### Pseudocode

```python
function kMeans(data, k):
    1. Initialize k cluster centroids randomly.
    2. Repeat until convergence:
        a. Assign each data point to the nearest centroid.
        b. Update each centroid as the mean of the data points assigned to it.
    3. Return the final cluster assignments and centroids.
```

````python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
````

## DataSet

````python
# Dataset 1
data1 = np.array([5, 8, 21, 6, 24, 23]).reshape(-1, 1)
# Dataset 2
data2 = np.array([(1, 2), (1, 3), (2, 2), (3, 2), (3, 4), (3, 3)])
````

## Defining The Cluster

````python
# Number of clusters
k = 2
````

## K-Means Clustering for the Abovementioned dataset 

````python
# K-means clustering for Dataset 1
kmeans1 = KMeans(n_clusters=k, random_state=42)
kmeans1.fit(data1)
labels1 = kmeans1.labels_
centroids1 = kmeans1.cluster_centers_

# K-means clustering for Dataset 2
kmeans2 = KMeans(n_clusters=k, random_state=42)
kmeans2.fit(data2)
labels2 = kmeans2.labels_
centroids2 = kmeans2.cluster_centers_
````

## Plotting the Dataset

````python
# Plotting for Dataset 1
plt.scatter(data1, np.zeros_like(data1), c=labels1, cmap='viridis')
plt.scatter(centroids1, np.zeros_like(centroids1), marker='X', s=200, color='red')
plt.title('K-means Clustering - Dataset 1')
plt.show()
````

````python
# Plotting for Dataset 2
plt.scatter(data2[:, 0], data2[:, 1], c=labels2, cmap='viridis')
plt.scatter(centroids2[:, 0], centroids2[:, 1], marker='X', s=200, color='red')
plt.title('K-means Clustering - Dataset 2')
plt.show()
````
